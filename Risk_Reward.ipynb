```python
%matplotlib inline
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

sns.set()
```

```python
ticker=["META","AMZN","NFLX","GOOG"]

stocks_data=yf.download(ticker,start="2025-01-01",end="2025-07-01")
```

```python
stocks_data.to_csv("stocks_data.csv")
```

```python
stocks=pd.read_csv("stocks_data.csv",header=[0,1],index_col=[0],parse_dates=[0])
stocks
```

```python
stocks.columns=stocks.columns.to_flat_index()
```

```python
stocks.columns
```

```python
stocks.columns=pd.MultiIndex.from_tuples(stocks.columns)
stocks
```

```python
close=stocks.loc[:,"Close"].copy()
close
```

```python
plt.figure(figsize=(15, 8))
plt.plot(close)
plt.title("Closing Price of Stocks (Jan–Jul 2025)", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Price (USD)", fontsize=12)
plt.legend(close.columns)
plt.show()
```

```python
#Normalizing the stocks price
norm_close=close.div(close.iloc[0]).mul(100)
```

```python
norm_close.plot(figsize=(15,8),fontsize=12)
plt.legend(fontsize=12)
plt.title("Normalized Price",fontsize=15)
plt.show()
```

```python
stock=stocks["Close"].copy()
stock
```

```python
ret=stock.pct_change().dropna()
```

```python
ret.plot(kind="hist",figsize=(12,8),bins=100)
plt.show()
```

```python
daily_mean=ret.mean()
daily_var=ret.var()
std=ret.std()
```

```python
annual_mean=daily_mean*252 
annual_var=daily_var*252
annual_std=ret.std()*np.sqrt(252)
```

```python
#Another way to get mean & std
summary=ret.describe().T.loc[:,["mean","std"]]
```

```python
summary["mean"]=summary["mean"]*252
summary["std"]=summary["std"]*np.sqrt(252)
```

```python
summary.plot.scatter(x="std",y="mean",figsize=(6,4),s=50,fontsize=11)
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,"std"]+0.002,summary.loc[i,"mean"]+0.002),size=13)
plt.xlabel("Annual Risk(std)",fontsize=15)
plt.ylabel("Annual Return( /100)",fontsize=15)
plt.title("Risk/Reward ",fontsize=15)
plt.show()
```
